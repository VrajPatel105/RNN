```python
import numpy as np

docs = [
'blue blue sky',
'cat cat sleeps',
'rain rain falls',
'code runs code',
'hello hello world',
'fast car fast',
'dog jumps dog',
'good day good',
'time time passes',
'red red rose',
'music plays music',
'shine bright shine',
'soft soft pillow',
'cold snow cold',
'light light shines',
'happy happy joy',
'run run run',
'dream dream dream',
'up up up we go',
'hot hot tea'
]
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer # had to use tf one too cuz idk why it was not allowing me to run keras.preprocessing.text
```

```python
tokenizer = Tokenizer(oov_token='<nothing>') # when during prediction, if there's some word that is not in our training data,
                                             # we use this concept where the new word that the model does not know, will be assigned as nothing
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.document_count # no. of rows
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

## we have to now apply padding, and this is because let's say we have one row that has about 3 words and the next one is 4 words.  


## Now since there's an imbalance, we can solve this by using padding where we add a zero either at the begenning of the vector or at the end of the vector that has missing values. So in this case, we can add one more zero at the end of the first row so that it equals out and turns into 2 rows with same amount of words.


hyperparameter:

                padding='post' -> padding at the end

                padding='pre' -> padding at first

```python
from keras.utils import pad_sequences
```

```python
sequences = pad_sequences(sequences, padding='post')
```

```python
sequences # here we can see that this is now padded data
```

# Now we can run the sentiment analysis main code

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(X_train, y_train),(X_test, y_test) = imdb.load_data()
```

we can see here that this data is already encoded with numbers, but we can see that there's no padding, so we can apply padding

```python
X_train = pad_sequences(X_train, padding='post')
X_test = pad_sequences(X_test, padding='post')
```

```python
len(X_train[0])
```

```python
model = Sequential()

model.add(SimpleRNN(32, input_shape=(len(X_train[0]),1), return_sequences=False)) # here the len(X_train[0]) is simply the time t. and 1 is the no. of outputs.
                                                        #return_sequences -> when we do forward propagation, each time the output of the last neuron is added to the next one but it this parameter is true, it will just print those output values as well.
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10, validation_data = (X_test,y_test))
```
